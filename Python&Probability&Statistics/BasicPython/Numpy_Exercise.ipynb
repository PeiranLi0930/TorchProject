```python
import numpy as np

x = np.array([1, 2, 3], dtype = np.float32) # 1 byte = 8 bits, 4 bytes = 32 bits

x, x.itemsize
```

```python
np.sin(x), np.tanh(x)
```

```python
x = np.ones((3, 3))
# add one dim
y = x[:, [0, 1, 2, 2]] # then y has independent memory area

# however, if z is generated by slicing x
z = x[:2, :2] # then z is just a view of x, so change the elements in x also change the elements in z

x[1, 1] = 9

y, z
```

```python
# Explicit force a copy
x = np.ones((3, 3))
z = x[:2, :2].copy() # z becomes independent

x[1, 1] = 9

x, z
```

```python
x.flags, "\n", z.flags # "flags" can help
```
